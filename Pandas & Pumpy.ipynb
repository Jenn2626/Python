1. Tạo 1 pandas series từ danh sách scores
2. Lọc tất cả điểm >=90
3. Tăng tất cả điểm thêm 5 và tạo 1 series mới
4. Tính tổng điểm và điểm tb
5. Tìm điểm < hơn điểm tb

```python
import pandas as pd
import numpy as np
scores =[85,95,78,64,89,95,73,88]
scores_series = pd.Series(scores)
scores_series
# Lọc tất cả điểm >=90
scores_series[scores_series >= 90]
```

```python
# Tăng tất cả điểm thêm 5 và tạo 1 series mới
new_scores_series = scores_series + 5
new_scores_series
```

```python
# Tính tổng điểm pandas
print('Tổng điểm:',new_scores_series.sum())
print('Tb điểm:',new_scores_series.mean())
```

```python
# Tính tổng điểm numpy
total = np.sum(new_scores_series)
mean = np.mean(new_scores_series)
print('Tổng điểm:',total)
print('Tb điểm:',mean)
```

```python
# Tìm điểm < hơn điểm tb
new_scores_series[new_scores_series < mean]
```
