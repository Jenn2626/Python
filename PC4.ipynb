#EX1: scores.csv
1.1 Sort and display teams in ascending order of score.

1.2 Sort and display teams in ascending order of category, then by score.

1.3 Calculate the mean of score for all teams.

1.4 Filter teams with score greater than 80 and display them in descending order of year.

1.5 Create a new column score_plus_10 by adding 10 to score.

1.6 Create a new column score_ratio as score / max(score).

1.7 Filter teams with category = ‘A’ or ‘B’ and score > 75.

1.8 Group by category and calculate the sum of score for each group.

```python
# Read file
import pandas as pd
df_scores = pd.read_csv('/content/scores.csv')
df_scores.head()
df_scores.shape
df_scores.info()
```

```python
# Sort and display teams in ascending order of score
df_scores.sort_values('score', ascending=True)
df_scores
```

```python
# Sort and display teams in ascending order of category, then by score
df_scores.sort_values(['category','score'],ascending=[True,True])
df_scores
```

```python
#  Calculate the mean of score for all teams
import numpy as np
df_scores['mean_score'] = np.mean(df_scores['score'])
df_scores
```

```python
# Filter teams with score greater than 80 and display them in descending order of year.

df_scores[df_scores['score'] > 80].sort_values('year', ascending=False)
```

```python
# Create a new column score_plus_10 by adding 10 to score.
df_scores['plus_score'] = df_scores['score'] + 10
df_scores
```

```python
# Create a new column score_ratio as score / max(score)
df_scores['score_ratio'] =  df_scores['score']/df_scores['score'].max()
df_scores
```

```python
# Filter teams with category = ‘A’ or ‘B’ and score > 75.
# df_scores['category'].isin(['A','B'])
# df_scores['score'] > 75
df_scores[(df_scores['category'].isin(['A','B'])) & (df_scores['score'] > 75)]
```

```python
# Group by category and calculate the sum of score for each group
df_scores.groupby('category')['score'].sum().reset_index()
df_scores
```

# EX2: euro_12.csv
2.1 Filter 5 columns: Team, Goals, Yellow Cards, Red Cards, Players Used.

2.2 Sort teams in ascending order of Players Used.

2.3 Sort teams in ascending order of Yellow Cards and Red Cards.

2.4 Calculate the mean of Yellow Cards for all teams.

2.5 Filter teams with more than 9 Goals and sort them in descending order of Passes.

2.6 Create a new column Shooting Accuracy = (Shots on target / (Shots on target + Shots off target)) × 100.

2.7 Sort and display teams, Goals, and Shooting Accuracy in descending order of Shooting Accuracy.

```python
# Read file
df_euro = pd.read_csv('/content/euro_12.csv')
df_euro.head()
df_euro.shape
df_euro.info()
```

```python
# Filter 5 columns: Team, Goals, Yellow Cards, Red Cards, Players Used.
df_euro[['Team','Goals','Yellow Cards','Red Cards','Players Used']].reset_index()
```

```python
# Sort teams in ascending order of Players Used.
sort_df = df_euro.sort_values('Players Used', ascending=True).reset_index()
sort_df
```

```python
#  Sort teams in ascending order of Yellow Cards and Red Cards.
sort_card = df_euro.sort_values(['Yellow Cards','Red Cards'], ascending=[True,True]).reset_index()
sort_card
```

```python
# Calculate the mean of Yellow Cards for all teams
import numpy as np
df_euro['mean_yellow'] = np.round((df_euro['Yellow Cards']).mean(),2)
df_euro
```

```python
# Filter teams with more than 9 Goals and sort them in descending order of Passes.
df_euro[df_euro['Goals'] > 9].sort_values('Passes', ascending=False)
```

```python
# Create a new column Shooting Accuracy = (Shots on target / (Shots on target + Shots off target)) × 100.
import numpy as np

df_euro['Shooting Accuracy'] = np.round(df_euro['Shots on target']/ (df_euro['Shots on target'] + df_euro['Shots off target'])*100,2)

df_euro
```

```python
# Sort and display teams, Goals, and Shooting Accuracy in descending order of Shooting Accuracy.

df_euro.sort_values(['Team', 'Goals', 'Shooting Accuracy'], ascending=[False,False,False])
df_euro
```

# EX3: users.txt
3.1 Calculate the mean age for each occupation.

3.2 Calculate the male ratio (%) for each occupation.

3.3 Sort male ratios in descending order.

3.4 Find and display min and max ages for each occupation.

3.5 Calculate and display mean ages of male and female for each occupation.

3.6 Count and display occupations by gender (male and female).

```python
# Create file
df_users = pd.read_csv('/content/users.txt', sep='|')
df_users.head()
df_users.shape
df_users.info()
```

```python
# Calculate the mean age for each occupation.
import numpy as np
df_users.groupby('occupation')['age'].mean().round(2).reset_index()
```

```python
# Calculate the male ratio (%) for each occupation.
import numpy as np

# Count total users per occupation
total_user = df_users.groupby("occupation")["gender"].count()

# Count male users per occupation
male_user = df_users[df_users["gender"] == "M"].groupby("occupation")["gender"].count()

# Calculate male ratio (%)
male_ratio = (male_user / total_user * 100).round(2).reset_index(name="male_ratio_percent")

male_ratio
```
