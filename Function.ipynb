Viet ham tinh chu vi va dien tich hinh chu nhat

```python
def tinh_cv_dt(d,r):
  cv = (d+r)*2
  dt = d*r
  return f"chu vi la: {cv}, dien tich la: {dt}"
tinh_cv_dt(5,6)
```

Viet ham de kiem tra xem co phai la so nguyen to khong. So nguyen to la so lon hon 1 va chi chia het cho chinh no


```python
def check_prime(number):
  if number < 2:
    return False
  else:
    for i in range(2,number):
      if number % i == 0:
        return False
      else:
        return True
check_prime(13)
```

Cho một danh sách nhiệt độ dưới dạng Celsius.
C = [0, 20, 25, 30, 37, 100]
Sử dụng map để chuyển đổi từng nhiệt độ sang Fahrenheit theo công thức: F = C *9/5 + 32.
Sử dụng filter để giữ lại các số > 70$.
Sử dụng reduce để tính trung bình của các nhiệt độ

```python
c_temp = [0, 20, 25, 30, 37, 100]

# Chuyen C sang F
f_temp = list(map(lambda x: x * 9/5 + 32,c_temp))

# Loc >70
filter_temp = list(filter(lambda x: x > 70,f_temp))
print(filter_temp)

# Tinh tb
from functools import reduce
tb_temp = reduce(lambda x,y: x+y,filter_temp)/len(filter_temp)
print(tb_temp)
```
