# Group by Challenge
1. Tính tổng doanh thu (price) cho từng sản phẩm (product_id).

2. Tính tổng doanh thu (price) theo cả product_id và seller_id.

3. Tính tổng và trung bình freight_value cho từng product_id.

4. Với mỗi người bán (seller_id), tính tổng price, trung bình freight_value, và đếm số đơn hàng (order_item_id).

```python
# Tính tổng doanh thu (price) cho từng sản phẩm (product_id)
"""1 hàm cho 1 cột """
df_order_items.groupby('product_id')['price'].sum().reset_index()
```

```python
# Tính tổng doanh thu (price) theo cả product_id và seller_id.
"""1 hàm cho nhiều cột """
df_order_items.groupby(['product_id','seller_id'])['price'].sum().reset_index()
```

```python
# Tính tổng và trung bình freight_value cho từng product_id.
"""Nhiều hàm cho nhiều (1) cột """
df_order_items.groupby('product_id').agg(total_freight_value=('freight_value','sum'),avg_freight_value=('freight_value','mean')).reset_index()

```

```python
#  Với mỗi người bán (seller_id), tính tổng price, trung bình freight_value, và đếm số đơn hàng (order_item_id).
"""Nhiều hàm và chỉ định từng hàm cho từng giá trị """
df_order_items.groupby('seller_id').agg( total_price = ('price','sum'),
                                        avg_freight_value = ('freight_value','mean'),
                                         count_orders = ('order_item_id','count')).reset_index()
```
